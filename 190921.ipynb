```python
import numpy as np
import os
import cv2
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import keras
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from keras.models import Sequential
from keras import optimizers
import matplotlib.pyplot as plt
%matplotlib inline
keras.__version__
```

```python
data_dir = "/Users/joannakang/Documents/Python_study/Kakao"
```

```python
class_name = os.listdir(data_dir)
print(class_name)
```

```python
Apeach = []
```

```python
add = os.path.join(data_dir, class_name[0])
```

```python
file = os.listdir(add)
print(file)
```

```python
range(len(file))
```

```python
for i in file:
    img = os.path.join(add, i)
    img = cv2.imread(img, cv2.IMREAD_COLOR)
    img = cv2.resize(img, (150, 150), interpolation = cv2.INTER_CUBIC)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.reshape((150, 150, 3))
    Apeach.append(img)
```

```python
plt.imshow(Apeach[0])
#plt.axis('off')  테두리없애기
plt.show()
```

```python
plt.figure(figsize= (15, 15)) #전체의 plot을 만드는 것
#서브플롯에 각각 데이터를 넣는다
plt.subplot(5, 5, 1) #서브플롯 사이즈를 지정한다 / 세번째 숫자는 0으로 넣으면 안됨...
plt.axis('off')
plt.imshow(Apeach[0])

plt.subplot(5, 5, 2)
plt.axis('off')
plt.imshow(Apeach[1])

plt.subplot(5, 5, 3)
plt.axis('off')
plt.imshow(Apeach[2])
plt.show()

#이렇게 하면 노가다...
```

```python
#반복문으로 중복되는 코드 줄이기
plt.figure(figsize =(15, 15))

for i in range(len(file)):
    plt.subplot(5, 5, i +1)
    plt.axis('off')
    plt.imshow(Apeach[i])
```
