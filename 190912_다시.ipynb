```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt
import numpy as np
import cv2
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from keras.models import Sequential
from keras import optimizers
%matplotlib inline
import os
```

```python
img = load_img('/Users/joannakang/Desktop/Emoticon/Apeach/1.jpg')
```

```python
img
```

```python
data = np.array(img)
samples = expand_dims(data,0)
datagen = ImageDataGenerator(brightness_range = [0.2, 10])
it = datagen.flow(samples, batch_size=1)
# generate samples and plot
for i in range(9):
    plt.subplot(330 + 1 + i)
    batch = it.next()
    image = batch[0].astype('uint8')
    plt.imshow(image)
plt.show()
```

```python
x = img_to_array(img) #이미지를 array로 바꿈
x = x.reshape((1,) + x.shape)
```

```python
i = 0
for batch in datagen.flow(x, batch_size= 32,
                          save_to_dir= '/Users/joannakang/Desktop/Emoticon/new', save_prefix='apeach', save_format='jpg'):
    i += 1
    if i > 50:
        break  # 이미지 50장을 생성하고 마칩니다
```
