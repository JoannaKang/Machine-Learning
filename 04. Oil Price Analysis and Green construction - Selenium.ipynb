```python
def scroll_and_read_text(make_xpath):
    variable = driver.find_elements_by_xpath(make_xpath)[0]
    driver.execute_script("return arguments[0].scrollIntoView();", variable)
    
    return(variable.text)
```

```python
def select_item(make_xpath):
    variable = driver.find_elements_by_xpath(make_xpath)[0]
    driver.execute_script("return arguments[0].scrollIntoView();", variable)
    time.sleep(0.1)
    ActionChains(driver).double_click(variable).perform()
    time.sleep(0.1)
```

```python
def get_ceri_date():
    make_xpath = """//*[@id="tab1c1"]/table[2]/tbody/tr[3]/td[1]"""
    return(scroll_and_read_text(make_xpath))
```

```python
def get_company_name():
    make_xpath = """//*[@id="tab1c3"]/div[1]/ul/li[1]"""
    return scroll_and_read_text(make_xpath)
```

```python
def get_homepage():
    make_xpath = """//*[@id="tab1c3"]/table/tbody/tr[5]/td[2]"""
    return scroll_and_read_text(make_xpath)
```

```python
def get_phone_num():
    make_xpath = """//*[@id="tab1c3"]/table/tbody/tr[3]/td[2]"""
    return scroll_and_read_text(make_xpath)
```

```python
def get_info():
    make_xpath = """//*[@id="tab1m1"]"""
    variable = driver.find_elements_by_xpath(make_xpath)[0].click()
    
    certi_date.append(get_ceri_date())

    make_xpath = """/html/body/div/ul/li[3]/a"""
    variable = driver.find_elements_by_xpath(make_xpath)[0].click()

    time.sleep(0.5)

    company.append(get_company_name())
    homepage.append(get_homepage())
    tell_no.append(get_phone_num())
```

```python
def switch_frame(frame_name):
    driver.switch_to_default_content()
    driver.switch_to_frame(frame_name)
```

```python
def error_procedure():
    company.append(np.nan)
    homepage.append(np.nan)
    tell_no.append(np.nan)
    certi_date.append(np.nan)
```

```python
from tqdm import tqdm_notebook

company = []
homepage = []
tell_no = []
certi_date = []

switch_frame('left')
```

```python
for list_name in tqdm_notebook(result_data['항목'][:100]):
    try:
        make_xpath = """//*[contains(text(), '""" + str(list_name) + """')]"""
        select_item(make_xpath)
        time.sleep(0.5)
        switch_frame('content')
        get_info()

    except:
        error_procedure()
        print(list_name, ' : --- error ---')
        
    switch_frame('left')
    time.sleep(1)
```
